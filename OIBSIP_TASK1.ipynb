<h1>IRIS PREDICTION</h1>
We are performing the classification by Logistic Regression Algorithm which is a classification algorithm and best suited for categorical data.

<h2>Importing libraries</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/kaggle/input/iris-oibsip/Iris.csv")
df
```

<h3>Let us see the datatypes and colums of the dataset</h3>

```python
df.info()
df.columns
```

```python
df.describe()
```

**Remove unwanted "ID" column**

```python
df=df.drop(columns="Id")
df
```

```python
df["Species"].value_counts()
```

Now split the columns into two variables x and y.
X being independent and y being dependent.

```python
x=df.iloc[:,:4]
y=df.iloc[:,4]
x
```

```python
y
```

Now split the data into Test and Train Datasets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

<h1>Create Model</h1>

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

**Accuracy of the model and Confusion Matrix**

```python
from sklearn.metrics import accuracy_score,confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```
